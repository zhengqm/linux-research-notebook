```python
import pandas as pd

# depends on SOB Graph
```

```python
data = pd.read_pickle("SOB.pickle")
```

```python
pairs = []
for k, v in data['undirected'].items():
    for t,w in v.items():
        pairs.append((k, t, w))
        
with open("SOB_undirected.edgelist", "w") as f:
    text = "\n".join(map(lambda p: str(p[0]) + ' ' + str(p[1]) + ' ' + str(p[2]), pairs))
    f.write(text)
```

```python
pairs = []
for k, v in data['directed'].items():
    for t,w in v.items():
        pairs.append((k, t, w))
        
with open("SOB_directed.edgelist", "w") as f:
    text = "\n".join(map(lambda p: str(p[0]) + ' ' + str(p[1]) + ' ' + str(p[2]), pairs))
    f.write(text)
```

```python
pairs = []
for k, v in data['directed'].items():
    for t,w in v.items():
        pairs.append((k, t, w))
        
with open("SOB_directed.edgelist.csv", "w") as f:
    f.write("Source,Target\n")
    text = "\n".join(map(lambda p: str(p[0]) + ',' + str(p[1]), pairs))
    f.write(text)

with open("SOB_directed.node.csv", "w") as f:
    f.write("id,label\n")
    text = "\n".join(map(lambda p: str(p[0]) + ',' + '"' + p[1].replace('"', '') + '"', data['id2email'].items()))
    f.write(text.encode("utf-8"))
    
```
