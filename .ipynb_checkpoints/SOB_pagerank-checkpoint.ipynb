```python
import pandas as pd
import networkx as nx
```

```python
result = {}

for year in [2011, 2012, 2013, 2014, 2015, 2016,2017]:
    for month in [1,7]:
        data = pd.read_pickle("SOB_{}_{}.pickle".format(year, month))
        links = [(k,v,w) for k, nodes in data['directed'].items() for v, w in nodes.items()]
        G = nx.DiGraph()
        G.add_weighted_edges_from(links)
        pr = nx.pagerank(G, max_iter=10000)
        pr_value = sorted([(v, k, data['id2email'][k]) for k,v in pr.items()], reverse=True)
        result[(year, month)] = pr_value
```

```python
mapping = {}

for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017]:
    for month in [1,7]:
        ranks = {t[1]: i for i, t in enumerate(result[(year, month)])}
        mapping[(year, month)] = ranks
```

```python
rank_df = pd.DataFrame(mapping)
```

```python
rank_df = rank_df.fillna(20000)
```

```python
rank_df.to_pickle("rank.pickle")
```
