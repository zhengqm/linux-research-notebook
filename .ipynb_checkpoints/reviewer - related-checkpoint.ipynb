```python
import pandas as pd
import re
from collections import defaultdict
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
commits = pd.read_pickle("commits.pkl")
```

```python
p1 = re.compile("Acked-by: (.*) <.*>")
p2 = re.compile("Reviewed-by: (.*) <.*>")
```

```python
df = commits[:1000]
```

```python
def reviewer_count(df):
    result = set()
    for msg in df.message:
        f1 = p1.findall(msg)
        f2 = p2.findall(msg)
        for dev in f1:
            result.add(dev)
        for dev in f2:
            result.add(dev)
    return len(result)
```

```python
result = {}
for year in range(2010, 2017):
    subset = commits[commits.committed_datetime.map(lambda t: t.year) == year]
    subset = subset[subset.message.map(lambda m: "Merge" not in m)]
    n_reviewers = reviewer_count(subset)
    n_authors = len(set(subset.author_name))
    n_commits = subset.shape[0]
    result[year] = {"reviewers": n_reviewers, "authors":n_authors, "commits": n_commits}
```

```python
df = pd.DataFrame(result).T
```

```python
plt.figure(facecolor="white")
(df.commits * 1.0 / df.reviewers).plot()
```

```python
plt.figure(facecolor="white")
(df.authors * 1.0 / df.reviewers).plot()
```

```python
same = commits[commits.author_name == commits.committer_name]
```

```python
same.shape[0], commits.shape[0], same.shape[0] * 1.0 / commits.shape[0]
```

```python
result = {}
for year in range(2010, 2017):
    subset = same[same.committed_datetime.map(lambda t: t.year) == year]
    checked = subset[subset.message.map(lambda m: "Reviewed-by:" in m or "Acked-by:" in m)]
    result[year] = {"total": subset.shape[0], "checked":checked.shape[0]}
```

```python
same_df = pd.DataFrame(result).T
same_df['ratio'] = 1.0 * same_df.checked / same_df.total
```

```python
plt.figure(facecolor="white")
same_df[['total', 'checked']].plot()
plt.show()
```

```python
plt.figure(facecolor="white")
same_df.ratio.plot()
```

```python
result = {}
for year in range(2010, 2017):
    subset = same[same.committed_datetime.map(lambda t: t.year) == year]
    diff = subset.committed_datetime - subset.authored_datetime
    result[year] = diff.describe()
```

```python
diff = pd.DataFrame(result).T
```

```python
diff['mean'].plot()
```

```python
diff['50%'].plot()
```

```python
result = {}
for year in range(2010, 2017):
    subset = same[same.committed_datetime.map(lambda t: t.year) == year]
    direct = subset[subset.authored_datetime == subset.committed_datetime]
    result[year] = {"total": subset.shape[0], "direct":direct.shape[0]}
```

```python
pd.DataFrame(result).T.plot()
```
