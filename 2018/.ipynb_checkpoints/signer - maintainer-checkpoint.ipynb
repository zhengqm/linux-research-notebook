```python
import pandas as pd
import re
from collections import Counter
%matplotlib inline
```

```python
commits = pd.read_pickle("../commits.pkl")
```

```python
pattern = re.compile("Signed-off-by: (.*) <.*>")
```

```python
pattern.findall(commits.message[12])
```

```python
# get top signers
```

```python
result = {}
for year in range(2008, 2018):
    selected = commits[commits.committed_datetime.map(lambda t: t.year == year)]
    print year, selected.iloc[0]['hexsha']
    counter = Counter()
    for msg in selected.message:
        signers = pattern.findall(msg)
        if len(signers) > 1:
            for signer in signers[1:]:
                counter[signer] += 1
    result[year] = counter.most_common(60)
```

```python
result
```

```python
m_pattern = re.compile('M:	"?(.*?)"? <.*>')
```

```python
maintainers = set()
for year in range(2008, 2018):
    with open("d:/linux/MAINTAINERS_{}".format(year)) as f:
        maintainer_text = f.read()
        maintainers = maintainers.union(set(m_pattern.findall(maintainer_text)))
```

```python
for year in range(2012, 2018):
    print year
    for name, _ in result[year]:
        if name not in maintainers:
            print name
```
