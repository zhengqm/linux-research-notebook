```python
import pandas as pd
import re
from collections import defaultdict
# depends on "Commit extraction"
```

```python
commit_df = pd.read_pickle("commits.pkl")
```

```python
pattern = "Signed-off-by: (.*) <.*>"
p = re.compile(pattern)
def insert(email, id2email, email2id):
    if email not in email2id:
        id = len(id2email)
        id2email[id] = email
        email2id[email] = id
    return email2id[email]
```

```python
id2email = {}
email2id = {}
        
for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017]:
    for (start, end) in [(1,6), (7, 12)]:
        subset = commit_df[commit_df.committed_datetime.map(lambda t: (t.year, t.month) >= (year, start) and (t.year, t.month) <= (year, end))]
        directed = defaultdict(lambda: defaultdict(int))
        undirected = defaultdict(lambda: defaultdict(int))


        for msg in subset.message:
            chains = p.findall(msg)
            for i in range(1, len(chains)):
                from_ = chains[i-1]
                to_ = chains[i]
                from_id = insert(from_, id2email, email2id)
                to_id = insert(to_, id2email, email2id)
                if from_id == to_id:
                    continue
                directed[from_id][to_id] += 1
                undirected[min(from_id, to_id)][max(from_id, to_id)] += 1
        data = {"directed": dict(directed), "undirected": dict(undirected), "email2id": email2id, "id2email": id2email}
        pd.to_pickle(data, 'SOB_{}_{}.pickle'.format(year, start))
```

```python
pairs = []
for k, v in undirected.items():
    for t in v:
        pairs.append((k, t))
        
with open("SOB_undirected.edgelist", "w") as f:
    text = "\n".join(map(lambda p: str(p[0]) + ' ' + str(p[1]), pairs))
    f.write(text)
```
