```python
import pandas as pd
from collections import defaultdict
```

```python
result = {}
for year in [2011, 2012, 2013, 2014, 2015, 2016]:
    for month in [1,7]:
        data = pd.read_pickle("SOB_{}_{}.pickle".format(year, month))
        edges = sum(w for _,v in data['directed'].items() for _, w in v.items())
        nodes = set()
        innodes = set()
        for src, v in data['directed'].items():
            for dst, w in v.items():
                nodes.add(src)
                nodes.add(dst)
                innodes.add(dst)
        result[(year, month)] = {'edges': edges, 'nodes': len(nodes), 'innode': len(innodes)}
```

```python
df = pd.DataFrame(result).T
```
