```python
import pandas as pd
```

```python
data = pd.read_pickle("SOB.pickle")
n_devs = len(data['id2email'])
indegree = pd.Series([0.0] * n_devs)
outdegree = pd.Series([0.0] * n_devs)

for k, vs in data['directed'].items():
    for v, w in vs.items():
        outdegree[k] += w
        indegree[v] += w
```

```python
indegree.sort_values(ascending=False)
```
