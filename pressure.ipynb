```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
commits = pd.read_pickle("commits.pkl")
```

```python
def get_sessions(timestamps):
    sessions = []
    start = timestamps.iloc[0]
    for i in range(1, len(timestamps)):
        delta = timestamps.iloc[i] - timestamps.iloc[i-1]
        if delta.seconds <= 30 * 60:
            continue
        else:
            last = timestamps.iloc[i-1]
            session = last - start
            start = timestamps.iloc[i]
            sessions.append(session)
    return pd.Series(sessions)
```

```python
result = {}
for year in range(2010, 2017):
    subset = commits[commits.committed_datetime.map(lambda t: t.year) == year]
    committers = subset.committer_name.value_counts().index
    arr = []
    for committer in committers:
        selected = subset[subset.committer_name == committer]
        timestamps = selected.committed_datetime.sort_values()
        sessions = get_sessions(timestamps)
        arr.append(sessions.max())
    result[year] = arr
```
