```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
commits = pd.read_pickle("commits.pkl")
```

```python
commits['has_tag'] = commits.message.map(lambda m: "Acked-by" in m or "Reviewed-by" in m or "Tested-by" in m )
```

```python
commits['has_review_tag'] = commits.message.map(lambda m: "Reviewed-by" in m)
commits['has_ack_tag'] = commits.message.map(lambda m: "Acked-by" in m )
commits['has_test_tag'] = commits.message.map(lambda m: "Tested-by" in m)
```

```python
g = commits.groupby(commits.committed_datetime.map(lambda t: t.year))
has_tag = {}
has_review_tag = {}
has_ack_tag = {}
has_test_tag = {}
for i, df in g:
    #print i
    has_tag[i] = df.has_tag.value_counts(normalize=True)
    has_review_tag[i] = df.has_review_tag.value_counts(normalize=True)
    has_ack_tag[i] = df.has_ack_tag.value_counts(normalize=True)
    has_test_tag[i] = df.has_test_tag.value_counts(normalize=True)
```

```python
plt.figure(facecolor="white")
pd.DataFrame(has_tag).T[True].plot()
```

```python
plt.figure(facecolor="white")
pd.DataFrame(has_review_tag).T[True].plot()
```

```python
plt.figure(facecolor="white")
pd.DataFrame(has_ack_tag).T[True].plot()
```

```python
plt.figure(facecolor="white")
pd.DataFrame(has_test_tag).T[True].plot()
```
