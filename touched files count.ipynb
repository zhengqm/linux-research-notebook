```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
commits = pd.read_pickle("commits.pkl")
```

```python
c = commits.files.map(lambda files: len(files.split(",")))
```

```python
c.quantile([0.9999])
```

```python
commits['count'] = c
```

```python
commits[commits['count'] > 230]
```

```python
def count_distinct_files(files_list):
    file_set = set()
    for files in files_list:
        if len(files) == 0:
            continue
        files = files.split(",")
        if len(files) > 230:
            continue
        for f in files:
            file_set.add(f)
    return len(file_set)
```

```python

result = {}
for year in range(2006, 2017):
    subset = commits[commits.committed_datetime.map(lambda t: t.year) == year]
    committers = subset.committer_name.value_counts()[:60].index
    arr = []
    for committer in committers:
        arr.append(count_distinct_files(subset[subset.committer_name == committer]['files']))
    result[year] = arr
```

```python
pd.DataFrame(result).describe()
```

```python
plt.figure(facecolor="white")
pd.DataFrame(result).median().plot()
```

```python
plt.figure(facecolor="white")
pd.DataFrame(result).mean().plot()
```
